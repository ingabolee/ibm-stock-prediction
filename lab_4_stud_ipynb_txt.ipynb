# Recurrent neural networks

# Task 1

Get IBM stock market data and evaluate prediction on a set of recurrent neural networks.

Dataset can be downloaded from https://www.kaggle.com/szrlee/stock-time-series-20050101-to-20171231



## Subtask 1

Load data and prepare the dataset

```python
import pandas as pd
import kagglehub

# Download latest version
path = kagglehub.dataset_download("szrlee/stock-time-series-20050101-to-20171231")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(f'{path}/IBM_2006-01-01_to_2018-01-01.csv')
df
```

```python
import matplotlib.pyplot as plt

plt.plot(df['Date'], df['Open'])
plt.show()
```

# Subtask 2

Perform prediction by using feed-forward neural network (sliding window)

```python
import numpy as np

def sliding_window(data, window_size):
  x, y = [], []
  for i in range(len(data) - window_size - 1):
    x.append(data[i:(i + window_size), 0])
    y.append(data[i + window_size, 0])

  return np.array(x), np.array(y)

df['Open'].fillna(df['Open'].mean(), inplace=True)
data = df['Open'].values.reshape(-1, 1)
print(f'{data} \n')

data = (data-data.min()) / (data.max() - data.min())
print(data)
window_size = 10

x, y = sliding_window(data, window_size)

split_index = int(len(x) * 0.8)
xtrain, xtest = x[:split_index], x[split_index:]
ytrain, ytest = y[:split_index], y[split_index:]
```

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(window_size,)),
    # Adding a Dense layer with 1 unit to match the target shape
    tf.keras.layers.Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='mse')
model.summary()

model.fit(xtrain, ytrain, epochs=200, batch_size=32)
```

```python
dense_loss = model.evaluate(xtest, ytest)
print(dense_loss)

predictions_dense = model.predict(xtest)
```

## Subtask 3

Perform prediction by using RNN

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam

model = tf.keras.models.Sequential([
    tf.keras.layers.SimpleRNN(10, activation='relu', input_shape=(window_size, 1)), # Using SimpleRNN layer
    tf.keras.layers.Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='mse')
model.summary()

model.fit(xtrain, ytrain, epochs=200, batch_size=32)
```

```python
rnn_loss = model.evaluate(xtest, ytest)
print(rnn_loss)

predictions_rnn = model.predict(xtest)
```

## Subtask 4

Perform prediction by using LSTM

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam

model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(10, activation='relu', input_shape=(window_size, 1)), # Using LSTM layer
    tf.keras.layers.Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='mse')
model.summary()

model.fit(xtrain, ytrain, epochs=200, batch_size=32)
```

```python
lstm_loss = model.evaluate(xtest, ytest)
print(lstm_loss)

predictions_lstm = model.predict(xtest)
```

Compare all above results by using MSE error and dermine the best architecture. You can do tests on more stock data than IBM.

```python
print('nn', dense_loss)
print('rnn', rnn_loss)
print('lstm', lstm_loss)

import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.plot(predictions_dense, label='Dense Predictions')
plt.plot(predictions_rnn, label='RNN Predictions')
plt.plot(predictions_lstm, label='LSTM Predictions')
plt.plot(ytest, label='Actual')
plt.legend()
plt.title('Actual vs. Predicted Values')
plt.xlabel('Time Steps')
plt.ylabel('Stock Price')
plt.show()
```
